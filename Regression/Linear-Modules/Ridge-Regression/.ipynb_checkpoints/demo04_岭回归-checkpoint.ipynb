# demo04_岭回归

```python
import numpy as np
import pandas as pd
import sklearn.linear_model as lm
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/Salary_Data2.csv')
x, y = data.loc[:, :'YearsExperience'], data['Salary']
```

```python
# 线性回归
model = lm.LinearRegression()
model.fit(x, y)
pred_y = model.predict(x)
```

```python
# 可视化
plt.figure('Linear Regression', facecolor='lightgray')
plt.title('Linear Regression', fontsize=22)
plt.grid(linestyle=':')
plt.scatter(x.values.ravel(), y, marker='s', s=80, color='dodgerblue',
           label='Samples')
plt.plot(x.values.ravel(), pred_y, color='orangered', 
         label='Regression Line')
plt.legend()
plt.show()
```

```python
# 找一些测试样本，看一下模型的评估结果
test_data = data[:30:4]
test_x, test_y = \
    test_data.loc[:, :'YearsExperience'], test_data['Salary']
# 预测
pred_test_y = model.predict(test_x)
# 评估模型误差
import sklearn.metrics as sm
# 评估模型的r2得分    取值区间：(-∞, 1] 分值越高，模型越好。
score = sm.r2_score(test_y, pred_test_y)
print("r2:", score)
```

```python
# 岭回归
model = lm.Ridge(100)
model.fit(x, y)
pred_y = model.predict(x)
# 可视化
plt.figure('Ridge Regression', facecolor='lightgray')
plt.title('Ridge Regression', fontsize=22)
plt.grid(linestyle=':')
plt.scatter(x.values.ravel(), y, marker='s', s=80, color='dodgerblue',
           label='Samples')
plt.plot(x.values.ravel(), pred_y, color='orangered', 
         label='Regression Line')
plt.legend()
plt.show()
```

```python
# 调参  根据测试样本的r2得分，把alpha调整到最佳值
params = np.arange(80, 130, 10)
for param in params:
    model = lm.Ridge(param)
    model.fit(x, y)
    pred_test_y = model.predict(test_x)
    score = sm.r2_score(test_y, pred_test_y)
    print('alpha:', param, '  r2:', score)
```
