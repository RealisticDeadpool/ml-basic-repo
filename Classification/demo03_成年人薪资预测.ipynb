# demo03_成年人薪资预测

```python
import numpy as np
import pandas as pd
```

```python
import sklearn.preprocessing as sp
data = pd.read_csv('../data/adult.txt', header=None)
data.head(3)
data[1] = sp.LabelEncoder().fit_transform(data[1])
data[3] = sp.LabelEncoder().fit_transform(data[3])
data[5] = sp.LabelEncoder().fit_transform(data[5])
data[6] = sp.LabelEncoder().fit_transform(data[6])
data[7] = sp.LabelEncoder().fit_transform(data[7])
data[8] = sp.LabelEncoder().fit_transform(data[8])
data[9] = sp.LabelEncoder().fit_transform(data[9])
data[13] = sp.LabelEncoder().fit_transform(data[13])
data[14] = sp.LabelEncoder().fit_transform(data[14])
data
```

```python
# 年龄、工作类型、序号、学历、工龄、婚姻、工作岗位、家庭身份、
# 人种、性别、投资收益、投资亏损、每周工作小时数、国籍、薪资水平
# data.info()
# data.describe()
# data[14].value_counts()
```

```python
import sklearn.ensemble as se
import sklearn.model_selection as ms

# 整理输入与输出
x, y = data.loc[:, :13], data[14]
# 创建模型
model = se.RandomForestClassifier(
    max_depth=9, n_estimators=140, random_state=7)
# 交叉验证
s = ms.cross_val_score(model, x, y, cv=5, scoring='f1_weighted')
s.mean()
```
